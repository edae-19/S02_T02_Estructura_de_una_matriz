```python
import numpy as np
```

```python
# EJERCICIO 1

numbers =  np.array ([1, 25, 4, 36, 12, 7, 9, 14, 12, 6, 5], dtype = 'i8')
print ("numbers es un array de {} dimensión con forma: {}".format(numbers.ndim, numbers.shape))
print ("Data type: {}".format (numbers.dtype))
```

```python
# EJERCICIO 2

meanNumbers = np.mean(numbers)
difference = 0

for number in numbers:
    difference += abs(number-meanNumbers)

meanDifference = difference / (numbers.size)

print ("La media de las diferencias es: {}".format(meanDifference))
```

```python
# EJERCICIO 3
matrix = np.array ([[1,7,8,14,9],
                   [12,1,9,8,0],
                   [6,24,7,9,46],
                   [4,4,8,7,9],
                   [3,2,5,9,1]])
maxValueMatrix = np.amax(matrix)
print ("El valor mas alto de la matriz es: {}".format(maxValueMatrix));

maxValueColumn = np.amax(matrix,axis = 0)
print ("Los valores mas altos de cada columna son: {}".format(maxValueColumn))

maxValueRow = np.amax(matrix,axis = 1)
print ("Los valores mas altos de cada fila son: {}".format(maxValueRow))
```

```python
# EJERCICIO 8

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
image = mpimg.imread('salto_angel.jpg')
imgplot = plt.imshow(image)
plt.show()
```

```python
#Eliminar el canal verde de la imagen [Red,Blue,Green]

moddified_image_NoGreen = image[:, :, [2,2,0]]
plt.imshow(moddified_image_NoGreen)

plt.imsave("salto_angel_NoGreen.png",moddified_image_NoGreen)
```
